## Onto Preprocessing the obtained Lyrics

```python
from bs4 import BeautifulSoup
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
def process(lyric):
   # lyric without HTML tags
   lyric = re.sub("\n", " ", lyric)
   # lyric without punctuation and numbers
   lyric = re.sub("[^a-zA-Z]",' ',lyric)
   # converting into lowercase and splitting to eliminate stopwords
   lyric = lyric.lower()
   lyric = lyric.split()
   # lyric without stopwords
   swords = set(stopwords.words("english"))  
   lyric = [w for w in lyric if w not in swords]               
   # joining of splitted paragraph by spaces and return
   return(" ".join(lyric))    
```

```python
lyrics_cleaned = []
for r in range(len(Lyrics)):
    if (r+1)%1000 == 0:
        print("No of reviews processed =", r+1)
    lyrics_cleaned.append(process(Lyrics[r]))
```

```python
moody['Lyrics_cleaned'] = lyrics_cleaned
```

### Separating the data into train and test datasets

```python
train_size = int(moody.shape[0] * 0.8)

train_sentences = moody.Lyrics_cleaned[:train_size]
train_labels = moody.Emotion[:train_size]

test_sentences = moody.Lyrics_cleaned[train_size:]
test_labels = moody.Emotion[train_size:]
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

max_len = 1000
```

```python
tokenizer = Tokenizer(num_words=num_words)
tokenizer.fit_on_texts(train_sentences)
```

### Creating the Embedding Matrix from the 400,000 words GloVe Dataset from Stanford

```python
embedding_dict = {}
with open("glove.6B.100d.txt", "r", encoding="utf8") as f:
    for line in f:
        values = line.split()
        word = values[0]
        vectors = np.asarray(values[1:], "float32")
        embedding_dict[word] = vectors
f.close()
```

```python
embedding_dict['the']
```

```python
num_words = len(word_index) + 1
embedding_matrix = np.zeros((num_words, 100))

for word, i in word_index.items():
    if i < num_words:
        emb_vec = embedding_dict.get(word)
        if emb_vec is not None:
            embedding_matrix[i] = emb_vec
```

#### One-Hot Encoding the Emotion (Target) Column

```python
train_labels[:5]
```

```python
train_labels = pd.get_dummies(train_labels).values
test_labels = pd.get_dummies(test_labels).values
```

## Onto the Model Making Part 

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense, Dropout,Bidirectional
from keras.initializers import Constant
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint
```

```python
import tensorflow as tf
tf.config.run_functions_eagerly(True)
```

```python
model = Sequential()
filepath = os.getcwd()
callback = ModelCheckpoint(filepath = filepath , monitor = 'val_accuracy' , save_best_only = True)

model.add(
    Embedding(
        num_words,
        100,
        embeddings_initializer=Constant(embedding_matrix),
        input_length=max_len,
        trainable=False,
    )
)
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(100)))
model.add(Dense(4, activation="sigmoid"))


optimizer = Adam(learning_rate=0.0006)

model.compile(loss="binary_crossentropy", optimizer=optimizer, metrics=["accuracy"])
```

```python
model.summary()
```

```python
history = model.fit(
    train_padded,
    train_labels,
    epochs=20,
    validation_data=(test_padded, test_labels),
    verbose=1,
    callbacks = [callback]
)
```

```python
model.save(os.getcwd())
```

```python
from tensorflow import keras
model = keras.models.load_model(os.getcwd())
```
